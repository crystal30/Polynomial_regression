# 多项式回归

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(low=-3.0, high=3.0, size=100)
y = 0.5 * (x**2) + x + 2 + np.random.normal(0, 1, size=100)
plt.scatter(x,y)
plt.show()
```

# 1.用线性回归进行拟合（在只有一个特征的情况下）

```python
from sklearn.linear_model import LinearRegression
Lin1 = LinearRegression()
X = x.reshape(-1,1)
Lin1.fit(X,y)
y_predict = Lin1.predict(X)
plt.scatter(X,y)
plt.plot(X, y_predict, color='r')
plt.show()
```

由上图,我们可看出,在一个特征下的线性回归不能很好的拟合所给出的样本点.   
所以,我们增加一个特征 x**2   

```python
X2 = np.hstack([X,X**2])
Lin2 = LinearRegression()
Lin2.fit(X2,y)
y2_predict = Lin2.predict(X2)
plt.scatter(X,y)
plt.scatter(X, y2_predict, color='r')
plt.show()
```

# 2.添加了一个特征 x**2 之后,拟合效果好了很多.

```python
plt.scatter(X,y)
plt.plot(X, y2_predict, color='r')
plt.show()
```

上图中是用折线图绘制的,很显然,是由于X的顺序错乱(因为np.random.uniform 产生的数据不是按从小到大进行排列的),导致有很多条折线,     
解决方法：（1）可以对X进行排序，（2）直接绘制散点图，一样可以看出来结果。  

```python
plt.scatter(X,y)
plt.plot(np.sort(x), y2_predict[np.argsort(x)], color='r')
plt.show()
```

```python
# 注意：上边我们用的是向量x来绘图   
# 下面,我们用矩阵X（只有一列）绘图，在np.sort()的用法中略有不同
plt.scatter(X,y)
plt.plot(np.sort(X,0), y2_predict[np.argsort(X,0)], color='r')
plt.show()
```
